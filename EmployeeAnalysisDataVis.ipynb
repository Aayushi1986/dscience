# JobRole Analysis

```python
a=pd.crosstab(df1.EmpJobRole,df1.PerformanceRating,margins='all')
a
```

```python
#SALES EXECUTIVES HAVE THE HIGHEST JOB PERFORMANCE AND TECHNICAL ARCHITECTS HAVE THE LOWEST.
```

```python
a1=pd.crosstab(df1.EmpJobRole,df1.PerformanceRating)
a1.plot(kind='bar', stacked=False, grid=True, title="JobRole Vs Performance")
```

# EducationBackground Analysis

```python
b=pd.crosstab(df1.EducationBackground,df1.PerformanceRating,margins='all')
b
```

```python
#EMPLOYEES WITH 'LIFE SCIENCES' BACKGROUND HAVE HIGHEST PERFORMANCE AND THOSE FROM 'HUMAN RESOURCES' HAVE THE LOWEST.
```

```python
b1=pd.crosstab(df1.EducationBackground,df1.PerformanceRating)
b1.plot(kind='bar', stacked=False, grid=True, title="Education Vs Performance")
```

# Department Wise Analysis

```python
c=pd.crosstab(df1.EmpDepartment,df1.PerformanceRating,margins="all")
c
```

```python
#THE DEPARTMENT 'SALES' HAS THE HIGHEST JOB PERFORMANCE WHILE THE DEPARTMENT 'DATA SCIENCE' HAS THE LOWEST.
#'DEVELOPMENT' HAS THE HIGHEST NUMBER OF JOB PERFORMERS.
```

```python
c1=pd.crosstab(df1.EmpDepartment,df1.PerformanceRating)
c1.plot(kind='bar', stacked=False, grid=True, title="EmpDepartment Vs Performance")
```

# Gender Wise Analysis

# OverTime and PerfromanceRating

```python
e=pd.crosstab(df1.OverTime,df1.PerformanceRating,margins="all")
e
```

```python
#EMPLOYEES WORKING OVERTIME HAVE A LOWER JOB PERFROMANCE RATING.
```

```python
e1=pd.crosstab(df1.OverTime,df1.PerformanceRating)
e1.plot(kind='bar', stacked=False, grid=True, title="Overtime Vs Performance")
```

# Age Wise Analysis

```python
f=pd.crosstab(df1.Age,df1.PerformanceRating, margins="all")
f
```

```python
#EMPLOYEES IN THE AGE-GROUP 34 TO 36 HAVE THE HIGHEST JOB PERFORMANCE.
```

```python
f1=pd.crosstab(df1.Age,df1.PerformanceRating)
p.plot(f1)

p.title("Age VS Performance")
```

# SUMMARY : BASED ON THE ABOVE DATA SALES EXECUTIVES AND THE SALES DEPARTMENT ARE GIVING A HIGHER JOB PERFORMANCE IN COMPARISON WITH THE OTHER EMPLOYEES. EMPLOYEES WORKING OVERTIME TEND TO HAVE A LOWER PERFORMANCE RATING. ALSO, EMPLOYEES WHO ARE TOO YOUNG AND TOO OLD HAVE A LOWER PERFORMANCE RATING.
